```python
# Del a)
# Denne koden leser inn en excelfil som heter "support_uke_24.xlsx". Denne ligger lagret i samme mappe som denne pythonkoden.

import pandas as pd  # Trenger denne for å lese inn excelfila
import numpy as np  # Bedre enn pandas på å håndere mye data effektivt.

# Leser excelfilen fra samme katalog som koden
data = pd.read_excel('support_uke_24.xlsx')

# Konverterer pandas DataFrame til NumPy-arrays
u_dag = data.iloc[:, 0].to_numpy()  # Data fra kolonne 1 - Ukedag i tabellen
kl_slett = data.iloc[:, 1].to_numpy()  # Data fra kolonne 2 - Klokkeslett i tabellen
varighet = data.iloc[:, 2].to_numpy()  # Data fra kolonne 3 - Varighet i tabellen
score = data.iloc[:, 3].to_numpy()  # Data fra kolonne 4 - Tilfredshet i tabellen

# Skriver ut resultater av array
print(type(u_dag))
print(u_dag)
print(kl_slett)
print(varighet)
print(score)

```

```python
# del b
# Her skal jeg lage ett søylediagram som viser antall hendelser for de 5 ukedagene
# Array er importer i oppgave a) og brukes videre her

import pandas as pd  # imporerer pandas (men den er også med i oppgave a)
import matplotlib.pyplot as plt  # importerer matplotlib.pyplot for å lage grafer

# teller antall henvendelser for de ulike dagene i uke 24
ukedag_antall = pd.Series(u_dag).value_counts()

# Sorter ukedagene i riktig rekkefølge
ukedager = ['Mandag', 'Tirsdag', 'Onsdag', 'Torsdag', 'Fredag']  # Forteller rekkefølgen jeg ønsker at dagene skal plottes i
ukedag_antall = ukedag_antall.reindex(ukedager).fillna(0)  # Gjør om på dataene mine og fyller eventuelle manglende verdier med 0

# Under her skal vi lage stolpediagrammet
plt.close('all')  # passer på å lukke eventuelle grafer

plt.figure(figsize=(12, 9))  # Angir hvor stor jeg vil stolpediagrammet skal være. 12 tommer bred, 9 tommer høy.

plt.bar(ukedager, ukedag_antall, width=0.9) # Lager som stolpediagram og setter en vidde på stolpene

# lage tekst på toppen av stolpene ved å bruke plt.txt()

offset_y = 0.5  # Definerer hvor mye over stolpene jeg vil antallet skal komme.

for k in range(len(ukedager)):  # Lager en løkke som går gjennom hver ukedag. k vil være indeksen for ukedagene
    plt.text(k, ukedag_antall[k] + offset_y, str(int(ukedag_antall[k])), ha = 'center')  # Plasserer teksten i senter på stolpene og gjør det om til tekst og heltall.


plt.xlabel('Ukedag')  # Forklarende tekst for hva som vises i x-aksen
plt.ylabel('Antall hendvendelser')  # Forklarende tekst for hva som vises i y-aksen
plt.title('Antall henvendelser fordelt på ukedager')  # Overskrift for stolpe-diagrammet
plt.grid(axis='y')  # Lage horisontale linjer for lettere å lese av
plt.savefig('Antall hendvendelser fordelt uke.pdf')  # lager en pdf fil på katalogen programmet ligger
plt.show()  # Få vist på skjerm
```

```python
# Del c)
# Nå vil jeg finne korteste og lengste samtaletid i fila og skrive det ut inni en tekst

korteste_tid = min(varighet)  # min gir laveste samtaletid i arrayen varighet - altså den korteste og lagrer i variabel korteste_tid
lengste_tid = max(varighet)  # max gir største samtaletid i arrayen varighet - altså den lengste og lagrer i variblen lengste_tid

# Benytter variabler for korteste og lengste samtaletid og skrive rdet ut i en informativ tekst
print(f'I uke 24 var den korteste samtaletiden på support {korteste_tid}, mens den lengste samtaletiden var {lengste_tid}')
```

```python
# Beregn gjennomsnittlig samtaletid  #lager en definisjon for å finne gjennomsnitt
from datetime import timedelta  # importeres for å kunne jobbe med tid

def dele_tid(tid_str):
    '''
    Denne funksjonen tar en tidsstreng i formatet 'HH:MM:SS' og konverterer den til et timedelta-objekt.
    '''
    h, m, s = map(int, tid_str.split(':'))  # siden klokkeslett er lagret slik 00:11:28 splittes de basert på ":" og tallene konverteres til heltall
    return timedelta(hours=h, minutes=m, seconds=s)  # returnerer et timedelta-objekt

def gjennomsnittlig_samtaletid(tid):
    '''
    Denne funksjonen tar en liste med tidsstrenger og beregner gjennomsnittlig samtaletid.
    '''
    total_samtaletid = sum((dele_tid(t) for t in tid), timedelta())  # summerer tiden
    gjennomsnittlig_tid = total_samtaletid / len(tid)  # finner gjennomsnittstiden
    return gjennomsnittlig_tid


# Lager variabelen gjennomsnittlig_tid ved å bruke funksjonen gjennomsnittlig_samtaletid
# Funksjonen tar arrayen varighet, som inneholder samtaletider, og beregner gjennomsnittlig samtaletid
gjennomsnittlig_tid = gjennomsnittlig_samtaletid(varighet)

# Konverter gjennomsnittlig tid til minutter
minutter = gjennomsnittlig_tid.total_seconds() // 60  # Beregner antall hele minutter
sekunder = gjennomsnittlig_tid.total_seconds() % 60  # Beregner sekundende


# Printer ut gjennomsnittlig samtaletid som tekst
print(f'Gjennomsnittlig samtaletid for alle samtalene er {minutter:.0f} minutter og {sekunder:.0f} sekunder.')


# Bruk help() for å få hjelp til å forstå hva funksjonene gjør
# help(dele_tid)  # viser dokumentasjonen for funksjonen dele_tid
# help(gjennomsnittlig_samtaletid)  # viser dokumentasjonen for funksjonen gjennomsnittlig_samtaletid
```

```python
# oppgave e

import matplotlib.pyplot as plt
from datetime import datetime

def skiftfordeling(tid_mottatt):
    '''
    Denne funksjonen tar en tidstreng i formatet 'HH:MM:SS' og returnerer hvilket skift tidsstrengen tilhører.
    '''

    tidspunkt = datetime.strptime(tid_mottatt, '%H:%M:%S')  # konverterer tidsstrengen til ett timedate-objekt

    # Definerer start og slutt for de ulike 2-timers skiftene
    skift1_start = datetime.strptime('08:00:00', '%H:%M:%S')
    skift1_slutt = datetime.strptime('10:00:00', '%H:%M:%S')
    skift2_start = datetime.strptime('10:00:01', '%H:%M:%S')
    skift2_slutt = datetime.strptime('12:00:00', '%H:%M:%S')
    skift3_start = datetime.strptime('12:00:00', '%H:%M:%S')
    skift3_slutt = datetime.strptime('14:00:00', '%H:%M:%S')
    skift4_start = datetime.strptime('14:00:00', '%H:%M:%S')
    skift4_slutt = datetime.strptime('16:00:00', '%H:%M:%S')

    # Bestemmer hvilket skift hendvendelsene tilhører basert på tidspunkt
    if skift1_start <= tidspunkt < skift1_slutt:
        return '8 - 10'
    elif skift2_start <= tidspunkt < skift2_slutt:
        return '10 - 12'
    elif skift3_start <= tidspunkt < skift3_slutt:
        return '12 - 14'
    else: 
        return '14 - 16'

# Kategoriser hver tid i kl_slett
skift_antall = {'8 - 10': 0, '10 - 12': 0, '12 - 14': 0, '14 - 16': 0}  # lager en ordbok med antall for de ulike skiftene
# En løkke som går gjennom hver tid som ligger i arrayen kl_slett. Så bruker den funksjonen skiftfordeling for å finne riktig skift og øke med 1.
for tid in kl_slett:
    skift = skiftfordeling(tid)
    skift_antall[skift] += 1

antall = skift_antall.values()  # Summering så jeg kan få antall i de ulike skiftene i kakediagrammet
labels = skift_antall.keys()  # Tekst til de ulike kakene

# Kakediagram
plt.close('all')
plt.figure(num=2, figsize=(12, 9))
plt.pie(antall, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#377eb8', '#4daf4a', '#984ea3', '#ff7f00'])  # Formaterer tekst til % og en desimal og gir de farge etter ønske
plt.title('Antall henvendelser per skift i uke 24')  # Overskrift på kakediagrammet
plt.axis('equal')  # Sørger for at kakediagrammet er sirkulært og ikke ovalt
plt.savefig('Antall_henvendelser_per_skift_uke_24_kake.pdf')  # Lagrer kakediagrammet til pdf-fil
plt.show()  # Viser kakediagrammet

           
```

```python
# oppgave f

import numpy as np  # Importerer biblioteket numpy for numeriske operasjoner

# Filtrer ut nan-verdier så vi bare har gyldige data
score = score[~np.isnan(score)]

# Fordel tilbakemeldinger inn i tre kategorier, negativ, nøytral og positiv
negative = score[(score >= 1) & (score <= 6)]  # De som har gitt tilbakemelding på score fra 1 - 6
noytral = score[(score >= 7) & (score <= 8)]  # De som har gitt tilbakemelding på score fra 7-8
positive = score[(score >= 9) & (score <= 10)]  # De som har gitt tilbakemelding på score fra 9-10

# Teller antall score i arrayen score i hver kategori ved hjelp av len
negative_antall = len(negative)
noytral_antall = len(noytral)
positive_antall = len(positive)

# Beregn prosentandel for positive og negative kunder
total_antall = len(score)  # Finner antallet som har gitt en tilbakemelding
positive_prosent = (positive_antall / total_antall) * 100  # Finner prosentenadelen for antall positive
negative_prosent = (negative_antall / total_antall) * 100  # Finner prosentenandelen for antall negative

# Beregn NPS - altså forskjellen mellom prosentandelene som er positive og negative.
nps = positive_prosent - negative_prosent

# Printer resultatene med forståelig tekst. Først antall, så prosent i parantes med to desimaler
print(f"Negative tilbakemeldinger: {negative_antall} ({negative_prosent:.2f}%)")
print(f"Nøytrale tilbakemeldinger: {noytral_antall}")
print(f"Positive tilbakemeldinger: {positive_antall} ({positive_prosent:.2f}%)")
print(f"Net Promoter Score (NPS): {nps:.0f}")
```
